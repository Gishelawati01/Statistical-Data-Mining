**Memahami Dataset SHARP**

T_REC adalah waktu dalam UT

HARPNUM adalah HMI Active Region Patch Number

NOAA_AR adalah NOAA Active Region

TOTUSJH	adalah total unsigned current helicity dalam G^2/m

TOTPOT	adalah total photospheric magnetic energy density in ergs per cubic centimeter

TOTUSJZ	adalah total unsigned vertical current, dalam Amperes

ABSNJZH	adalah absolute value of the net current helicity in G^2/m

SAVNCPP adalah sum of the Absolute Value of the Net Currents Per Polarity in Amperes

USFLUX adalah total unsigned flux in Maxwells

MEANPOT adalah mean photospheric excess magnetic energy density in ergs per cubic centimeter

SHRGT45	adalah percentage of pixels with a mean shear angle greater than 45 degrees in percent

MEANSHR	adalah mean shear angle (measured using B_total) in degrees

MEANGAM	adalah Mean inclination angle, gamma, in degrees

MEANGBT	adalah Mean value of the total field gradient, in Gauss/Mm

MEANGBZ	adalah Mean value of the vertical field gradient, in Gauss/Mm

MEANGBH	adalah Mean value of the horizontal field gradient, in Gauss/Mm

MEANJZH	adalah Mean current helicity in G^2/m

MEANJZD	adalah Mean vertical current density, in mA/m^2

MEANALP	adalah Total twist parameter, alpha, in 1/Mm

**Pre processing Data**

```python
# Mengganti nilai kosong pada kolom CLASS dengan "No Flare"

data['CLASS'] = data['CLASS'].fillna('No Flare')
data

# Kami mengganti nilai kosong tersebut dengan No Flare karena :
# 1. flare tidak terjadi setiap saat, sehingga sangat mungkin tidak terjadi flare pada saat sedang diamati
# 2. Karena flare tidak teramati, maka tidak teridentifikasi kelasnya
# 3. Kolom CLASS ini penting jika ingin menganalisis flare terhadap deret waktu atau untuk machine learning
# Oleh karena itu kami menggantinya dengan "No Flare"
```

```python
# Mengganti nilai Mengganti nilai inf dengan NaN
pd.set_option('use_inf_as_na', True)
#Dari data statistik, terlihat bahawa terdapat kolom/feature yang nemiliki nilai minimum dan maksimum yang inf.
#Sehingga perlu dilakukan pembatasan pada rentang nilai tertentu.
```

```python
# Menghapus baris yang memiliki data kosong

data = data.dropna()
data

# Kami menghapus baris yang memiliki data kosong karena :
# 1. Jumlah baris kosong pada kolom MEANPOT(7419), R_VALUE(3964), SHRGT45(7419), MEANSHR(9615), MEANGAM(7939)
# MEANGBT(7427), MEANGBZ(7419), MEANGBH(7419), MEANJZH(7423), MEANJZD(7419), MEANALP(7427) sedikit dibandingkan jumlah data keseluruhan (2111903)
# 2. Jika menghapusnya, informasi penting masih bisa didapat dengan dataset yang ada

```

```python
# Mengecek keberhasilan menghapus data kosong

data.isnull().any()
```
